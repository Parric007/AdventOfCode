# Day 4 A1

```python
inputDF = pd.read_csv("Day4Input.txt", sep="|")

#print(inputDF)
def getNumberArray(inputArr):
    return inputArr.split(" ")
def getNumberArrayWin(inputArr):
    return inputArr.split(":")[1].split(" ")

inputDF["einzelGewinn"] = inputDF["Gewinnzahlen"].apply(getNumberArrayWin)
inputDF["einzelZahlen"] = inputDF["Zahlen"].apply(getNumberArray)

#print(inputDF["einzelGewinn"][0])
result = 0
for i in range (len(inputDF)): 
    internalResult = 0
    for j in range (len(inputDF["einzelGewinn"][i])):
        for k in range (len(inputDF["einzelZahlen"][i])):
            if(inputDF["einzelGewinn"][i][j] == inputDF["einzelZahlen"][i][k] and inputDF["einzelZahlen"][i][k] != ""):
                #print(inputDF["einzelGewinn"][i][j], inputDF["einzelZahlen"][i][k])
                if(internalResult == 0):
                    internalResult = 1
                else:
                    internalResult *= 2
    #print(internalResult)
    result += internalResult
print(result)
```

```python
inputDF = pd.read_csv("Day4Input.txt", sep="|")

#print(inputDF)
def getNumberArray(inputArr):
    return inputArr.split(" ")
def getNumberArrayWin(inputArr):
    return inputArr.split(":")[1].split(" ")

inputDF["einzelGewinn"] = inputDF["Gewinnzahlen"].apply(getNumberArrayWin)
inputDF["einzelZahlen"] = inputDF["Zahlen"].apply(getNumberArray)

#print(inputDF["einzelGewinn"])
#print(inputDF["einzelZahlen"])
#print(inputDF)


result = 0
i = 0
while( i < len(inputDF)):
    internalResult = 0
    for j in range (len(inputDF["einzelGewinn"][i])):
        for k in range (len(inputDF["einzelZahlen"][i])):
            if(inputDF["einzelGewinn"][i][j] == inputDF["einzelZahlen"][i][k] and inputDF["einzelZahlen"][i][k] != ""):
                #print(inputDF["einzelGewinn"][i][j], inputDF["einzelZahlen"][i][k])
                internalResult+=1
    #print(internalResult)
    for l in range(1,internalResult+1):
        #print(l)
        inputDF = pd.concat([inputDF, pd.DataFrame([[0, 0, inputDF["einzelGewinn"][i+l],inputDF["einzelZahlen"][i+l]]], columns=inputDF.columns)], ignore_index=True)
    inputDF.reset_index()
    i+=1
    if(i%1000 == 0):
        print(i)
print(len(inputDF))


#11787590
```
