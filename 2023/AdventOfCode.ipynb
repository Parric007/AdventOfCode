```python
import pandas as pd
import numpy as np
```

```python
result= 0
def calcResult(s : str):
    s = s.replace("one", "o1e").replace("two", "t2o").replace("three", "t3e") \
            .replace("four", "f4r").replace("five", "f5e").replace("six", "s6x") \
            .replace("seven", "s7n").replace("eight", "e8t").replace("nine", "n9e")
    first = np.NaN
    for c in s:
        try: 
            last = int(c)
            if np.isnan(first):
                first = last
                
        except:
            pass

    return first*10+last

text = pd.read_csv("Day1Input.txt")
text["val"] = text["Titel"].apply(calcResult)

for val in text["val"]:
    result += val
    
print(result)
```

```python
redCubes = 12
greenCubes = 13
blueCubes = 14

def extractCubes(s : str):
    game = s.split(";")
    maxNumber = {"red": 0, "green": 0, "blue": 0}
    for r in game:
        count = {"red": 0, "green": 0, "blue": 0}
        toCalcGame = r.split(",")
        for c in toCalcGame:
            c = c.split(" ")
            count[c[2]] += int(c[1])
        if(count["red"] > maxNumber["red"]):
            maxNumber["red"] = count["red"]
        if(count["green"] > maxNumber["green"]):
            maxNumber["green"] = count["green"]
        if(count["blue"] > maxNumber["blue"]):
            maxNumber["blue"] = count["blue"]
            
    if(maxNumber["red"] == 0):
        maxNumber["red"] = 1
    if(maxNumber["green"] == 0):
        maxNumber["green"] = 1
    if(maxNumber["blue"] == 0):
        maxNumber["blue"] = 1
    
        
    return maxNumber["red"]*maxNumber["green"]*maxNumber["blue"]
        

text = pd.read_csv("Day2Input.txt", sep=":")
text["Cubes"] = text["Rounds"].apply(extractCubes)

#resDF = text.loc[text["Cubes"] == "Alles da"]

result = 0

for strings in text["Cubes"]:
    result += int(strings)

print(result)
```
