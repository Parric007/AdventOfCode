```python
import pandas as pd
import numpy as np
import re
from typing import Generator, Tuple, List
from IPython.display import clear_output
import time
```

```python
result= 0
def calcResult(s : str):
    s = s.replace("one", "o1e").replace("two", "t2o").replace("three", "t3e") \
            .replace("four", "f4r").replace("five", "f5e").replace("six", "s6x") \
            .replace("seven", "s7n").replace("eight", "e8t").replace("nine", "n9e")
    first = np.NaN
    for c in s:
        try: 
            last = int(c)
            if np.isnan(first):
                first = last
                
        except:
            pass

    return first*10+last

text = pd.read_csv("Day1Input.txt")
text["val"] = text["Titel"].apply(calcResult)

for val in text["val"]:
    result += val
    
print(result)
```

```python
redCubes = 12
greenCubes = 13
blueCubes = 14

def extractCubes(s : str):
    game = s.split(";")
    maxNumber = {"red": 0, "green": 0, "blue": 0}
    for r in game:
        count = {"red": 0, "green": 0, "blue": 0}
        toCalcGame = r.split(",")
        for c in toCalcGame:
            c = c.split(" ")
            count[c[2]] += int(c[1])
        if(count["red"] > maxNumber["red"]):
            maxNumber["red"] = count["red"]
        if(count["green"] > maxNumber["green"]):
            maxNumber["green"] = count["green"]
        if(count["blue"] > maxNumber["blue"]):
            maxNumber["blue"] = count["blue"]
            
    if(maxNumber["red"] == 0):
        maxNumber["red"] = 1
    if(maxNumber["green"] == 0):
        maxNumber["green"] = 1
    if(maxNumber["blue"] == 0):
        maxNumber["blue"] = 1
    
        
    return maxNumber["red"]*maxNumber["green"]*maxNumber["blue"]
        

text = pd.read_csv("Day2Input.txt", sep=":")
text["Cubes"] = text["Rounds"].apply(extractCubes)

#resDF = text.loc[text["Cubes"] == "Alles da"]

result = 0

for strings in text["Cubes"]:
    result += int(strings)

print(result)
```

```python
clear_output(wait=True)
inputText = pd.read_csv("Day3Input.txt")
startTime = time.perf_counter()
inputMatrix = []
for i in inputText["Parts"]:
    inputMatrix.append(list(i))
    
internalCounter = 0
result1 = 0

for i in range (len(inputMatrix[0])):
    for j in range (len(inputMatrix)):
        if(internalCounter > 0):
            internalCounter -= 1
            continue
        internalResult = 0
        internalCounter = 0
        isSymbolAdjacent = False
        try:
            while inputMatrix[i][j+internalCounter].isdigit():
                internalCounter += 1
            if(internalCounter != 0):
                for k in range (internalCounter-1,-1,-1):
                    internalResult += int(inputMatrix[i][j+k])*(10**(internalCounter-k-1))
            for h in range(internalCounter):
                if isSymbolAdjacent:
                    break
                for k in range(-1,2):
                    for m in range(-1,2):
                        currentChar = inputMatrix[i+k][j+m+h]
                        if(isSymbol(currentChar)):
                            isSymbolAdjacent = True
                            break
                            
            if isSymbolAdjacent:
                result1 += internalResult
        except:
            continue
print(result1)
print(f"In Time: {time.perf_counter() - startTime:0.4f} seconds")


#print(gearPositions)                
```

```python
clear_output(wait=True)
inputText = pd.read_csv("Day3Input.txt")

startTime2 = time.perf_counter()

inputMatrix = []
for i in inputText["Parts"]:
    inputMatrix.append(list(i))
    

def isSymbol(st : str):
    return st != "." and not st.isdigit()
def isStar(st : str):
    return st == "*"

def getIndex(arr, tupel):
    if(len(arr) != 0):
        for a in arr:
            try:
                checkTupel = a[0]
            except:
                continue
            if(checkTupel == tupel):
                return arr.index(a)
    return -1

gearPositions = [[]]
result2 = 0
internalCounter = 0

for i in range (len(inputMatrix[0])):
    for j in range (len(inputMatrix)):
        if(internalCounter > 0):
            internalCounter -= 1
            continue
        internalResult = 0
        internalResultMult = 0
        internalCounter = 0
        foundMatch = False
        while inputMatrix[i][j+internalCounter].isdigit():
            internalCounter += 1
        if(internalCounter != 0):
            for k in range (internalCounter-1,-1,-1):
                internalResult += int(inputMatrix[i][j+k])*(10**(internalCounter-k-1))
        for h in range(internalCounter):
            if foundMatch:
                break
            for k in range(-1,2):
                for m in range(-1,2):
                    currentChar = inputMatrix[i+k][j+m+h]
                    if(isStar(currentChar)):
                        ind = getIndex(gearPositions, (i+k,j+m+h))
                        if (ind != -1):
                            if(gearPositions[ind][1] != internalResult):
                                internalResultMult = gearPositions[ind][1] * internalResult
                                foundMatch = True
                        else:
                            gearPositions.append([(i+k,j+m+h), internalResult])

                        break
                            
            if foundMatch:
                result2 += internalResultMult

print(result2)
print(f"In Time: {(time.perf_counter() - startTime2)*1000:0.4f} milliseconds")
```
