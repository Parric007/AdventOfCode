```python
#12357789728873
ll = [x for x in open("Day8Input.txt").read().strip().split("\n\n")]

instructions = list(ll[0])
conn = {}

for l in ll[1].split("\n"):
    current = l.split(" ")[0]
    leftTarget = l.split("(")[1].split(",")[0]
    rightTarget = l.split(" ")[3].split(")")[0]
    conn[current] = (leftTarget, rightTarget)
    
pos = "AAA"
index = 0
while pos != 'ZZZ':
    d = instructions[index%len(instructions)]
    pos = conn[pos][0 if d=="L" else 1]
    index += 1

def solvesteps(start):
    pos = start
    index = 0
    while not pos.endswith("Z"):
        d = instructions[index%len(instructions)]
        pos = conn[pos][0 if d=="L" else 1]
        index += 1
    return index

result = 1
for start in conn:
    if start.endswith("A"):
        result = math.lcm(result, solvesteps(start))
print(result)
#print(conn)
```
